```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange', lw=2, ls='--')
ax.set_title='My Title'
ax.set_xlabel='My x'
ax.set_ylabel='My y'
ax.plot(x,y,color='blue')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,0.5,0.2,0.2]) # inset axes
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1]) # main axes
ax2 = fig.add_axes([0.2,0.5,0.2,0.2]) # inset axes
ax1.plot(x,y,color='blue')
ax2.plot(x,y,color='red')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])  #main
axes2=fig.add_axes([0.2,0.5,0.4,0.4])

x = np.arange(0,100)
z=x**2
y=x*2

axes1.plot(x,z,label = 'x sqd')
axes2.plot(x,y,label = '2x')

axes1.set_xlim([0,100])
axes2.set_xlim([20, 22])
axes2.set_ylim([30, 50])

axes2.set_title('zoom')

axes1.set_xlabel('X Label')
axes1.set_ylabel('Z Label')

axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')


```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
x = np.arange(0,100)
y=x*2
z=x**2
axes[0].set_title('Graph 0')
axes[1].set_title('Graph 1')
axes[0].set_xlim([0,100])
axes[0].set_xlabel('x')
axes[1].set_xlabel('x')
axes[0].plot(x,y,color='blue',lw=2,ls="--")
axes[1].plot(x,z,color='red',linewidth=2,ls="dashdot")


```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
fig.set_figheight(4)
fig.set_figwidth(18)

x = np.arange(0,100)
y=x*2
z=x**2
axes[0].set_title('Graph 0')
axes[1].set_title('Graph 1')
axes[0].set_xlim([0,100])
axes[0].set_xlabel('x')
axes[1].set_xlabel('x')
axes[0].plot(x,y,color='blue',lw=2,ls="--")
axes[1].plot(x,z,color='red',linewidth=2,ls="dashdot")
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")
fig.subplots_adjust(left=0.15, right=.9, bottom=0.1, top=0.9);
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```
