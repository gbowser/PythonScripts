```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import openpyxl as xl
import scipy

from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)

%matplotlib inline

from openpyxl import load_workbook

from scipy.interpolate import make_interp_spline, BSpline

```

```python

path = "d:\\Dropbox\\Public Documents\\Astronomy\\UCLAN\\AA1055 IT for Astronomy\\Assignment 3\\FITS 0018.xlsm"

wb_obj = xl.load_workbook(path)

# Get the Line Fitting Novas , put in DF1
ws1 = wb_obj['Pygraph']
data = ws1.values
# Get the first line in file as a header line
columns = next(data)[0:]
# Create a DataFrame based on the second and subsequent lines of data
df1 = pd.DataFrame(data, columns=columns)
```

```python
print (df1)
```

```python
x1=df1['HJD']
y1=df1['Mag']

yerror=df1['Error']
```

```python

xnew = np.linspace(min(x1),max(x1), 500) 
spl = scipy.interpolate.make_interp_spline(x1, y1, k=5)
y_smooth = spl(xnew)

plt.plot(xnew,y_smooth)


fig, ax = plt.subplots()
ax.scatter(x1, y1, marker='o', s=20.0,  edgecolors='k')
plt.grid(b=True, which='major', axis='both')
ax.errorbar(x1,y1,yerr=yerror, linestyle="None", elinewidth=1, capsize=4,color='k')


ax.set_ylim(-6.0,-10.0);


# fix small tick labels & graph size
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))
fig = plt.gcf()


ax.set_xlabel("Heliocentric Julian Date",fontsize=14)
ax.set_ylabel("Instrumental Magnitude",fontsize=14)

plt.title('Light Curve of Stars');
#  fig.set_size_inches(8,8)

```
